```python
data= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CSVs/bank-loan.csv')

# create a dataframe with the given data. Keep the original data data in variable data (df is the common abbreviation for dataframes)
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df.age,df.income)
plt.show()
```

```python

# size of figure
fig, ax = plt.subplots(figsize=(12, 5))

# style of figure
sns.set_style("darkgrid")

# set title of figure
ax.set_title("employment durations by age")

# adjust default plot style
sns.set_context("notebook",font_scale=1.3)

# create barplot form library seaborn
sns.barplot(x='age', y='employ',data=df, palette='Blues', errorbar=None)

# set names of axes
ax.set_xlabel("age")
ax.set_ylabel("employment_duration")

plt.show()
```

```python
# select only numerical values
numericals = df.select_dtypes(include='number')

# corr of numerical values
corr_numericals = numericals.corr()
corr_numericals= round(corr_numericals, 2)
# plot heatmap
plt.figure(figsize=(12,6))
# abs is just another way to visualize the corr, by absolute values
sns.heatmap(corr_numericals, cmap='Blues', annot=True)
plt.show()
```

```python
#histogramm with plt

plt.hist(df.creddebt)
```

```python
# loop over all numerical columns and create histograms of their distributions
for column in df.select_dtypes(include='number').columns:

    sns.set(style="darkgrid")
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
    plt.tight_layout()
    plt.show()
```
